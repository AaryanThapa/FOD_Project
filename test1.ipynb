```python
import numpy as np

A = np.array([[2, -1],
              [0,  3]])

B = np.array([[ 1,  2],
              [-3,  4]])

```

```python
sum = A + B
print("A+B=")
print(sum)

```

```python
AB=A@B
print("AB=")
print(AB)
```

```python
result=3*A-2*B
print("3*A-2*B=")
print(result)
```

```python
determinant= np.linalg.det(A)
print("Det of A=")
print(determinant)
```

Here determinant of A is not zero hence it is invertible.

```python
inv=np.linalg.inv(A)
print("A⁻¹ =")
print(np.round(inv,3))
```
