## Findings from line chart(Age vs SYS Blood pressure)
-Here the line chart shows how SYS blood pressure varies with age for different person.
-Here lot of fluctuating lines are seen because many peoples of same age have different Blood pressure.
-This graph gives a relation between age and SYS blood pressure.


```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("Health_heart_experimental.csv")


df_sorted = df.sort_values("age")

plt.plot(df_sorted["age"], df_sorted["SysBP"])
plt.xlabel("Age")
plt.ylabel("Sys BP")
plt.title("Age vs Sys Blood Pressure")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Health_heart_experimental.csv")

bmi = df.groupby("sex")["BMI"].mean()

plt.bar(["Female", "Male"], bmi, color=['Red', 'blue'])
plt.title("BMI by Sex")
plt.ylabel("BMI Level")
plt.show()
```
