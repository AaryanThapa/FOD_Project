```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("Health_heart_experimental.csv")


df_sorted = df.sort_values("age")

plt.plot(df_sorted["age"], df_sorted["SysBP"])
plt.xlabel("Age")
plt.ylabel("Sys BP")
plt.title("Age vs Sys Blood Pressure")
plt.show()
```

## Findings from line chart(Age vs SYS Blood pressure)
-Here the line chart shows how SYS blood pressure varies with age for different person.
-Here lot of fluctuating lines are seen because many peoples of same age have different Blood pressure.
-This graph gives a relation between age and SYS blood pressure.


```python
import matplotlib.pyplot as plt
df = pd.read_csv("Health_heart_experimental.csv")
bmi = df.groupby("sex")["BMI"].mean()


plt.bar(["Female", "Male"], bmi, color=['red', 'blue'])
plt.title("BMI by Sex")
plt.ylabel("BMI")
plt.show()
```

## Findings from Bar PLot(BMI by Sex )
-This Bar plot shows the average body mass index from males and females from given dataset.
-This plot help us to comapre overall BMI between two sexes.

```python
df['indication'].value_counts().plot(kind='pie',autopct='%0.2f')
```

## FIndings from pie chart
-This pie chart distribution of the indication column in the dataset.
-The `autopct='%0.2f'` shows the exact percentage of each group on the chart. 
-Each slice represents the proportion of people with indication = 0 (No Indication) and indication = 1 (Indication).
